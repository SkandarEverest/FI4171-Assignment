```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def montecarlo(N,r,a=0,b=0):

    x = []
    y = []
    iterasi = []
    error = []
    hasil = np.pi * (r**2)
    print(hasil)
    
    
    x.append(np.random.uniform(low=-1, high=1))
    y.append(np.random.uniform(low=-1, high=1))
    iterasi.append(1)
    
    xi=np.array(x)
    yi=np.array(y)

    inside = (xi**2 + yi**2) <(r**2)

    pi_aproksimasi = 4*np.sum(inside)
    luaslingkaran = pi_aproksimasi*(r**2)
    error.append(abs(luaslingkaran - hasil)/hasil*100)
    
    for i in range(1,N):
        unique1 = 0
        unique2 = 0
        while(unique1 != 1):
            unique1 = 1
            xtmp = np.random.uniform(low = ((-1)*r),high = r)
            for j in range(i):
                if(xtmp == x[j]):
                    unique1 = 0
                    break
            x.append(xtmp)
             
        while(unique2 != 1):
            unique2 = 1
            ytmp = np.random.uniform(low = ((-1)*r),high = r)
            for k in range(i):
                if(ytmp == y[k]):
                    unique2 = 0
                    break
            y.append(ytmp)
    
        xi=np.array(x)
        yi=np.array(y)

        inside = (xi**2 + yi**2) <(r**2)

        pi_aproksimasi = 4*np.sum(inside)/(i+1)
        #print(pi_aproksimasi,i)
        luaslingkaran = pi_aproksimasi*(r**2)
        #print(luaslingkaran)
        iterasi.append(i+1)
        error.append(abs(luaslingkaran - hasil)/hasil*100)

    #print(iterasi)
    #print(error)
    x_in = xi[inside]
    y_in = yi[inside]

    plt.figure(figsize = [5,5])
    plt.scatter(xi+a,yi+b,s=1)
    plt.scatter(x_in+a,y_in+b,color = 'r',s=1)
    plt.show()
    
    plt.plot(iterasi,error)
    plt.xlabel('Iterasi')
    plt.ylabel('Error (%)')
    plt.show
```

```python
montecarlo(10000,5)
```
