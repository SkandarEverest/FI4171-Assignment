```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import math
```

```python
dt = 0.1
m = 1 #float(input('Massa = '))
g = 9.8
N = 28 #Partisi
v = np.zeros(N+1)
y = np.zeros(N+1)
t = np.zeros(N+1) 
ek = np.zeros(N+1)
ep = np.zeros(N+1)
em = np.zeros(N+1)
```

```python
t[0] = 0.0
y[0] = 0 #float(input('Posisi awal = '))
vy = 20 #float(input('Kecepatan awal = '))
sudut = math.radians(45)  #int(input('Sudut = ')))
```

```python
def energi_potensial(h):
    return (m*g*h)
def energi_kinetik(v):
    return (1/2*m*(v**2))
def energi_mekanik(kin,pot):
    return (kin + pot)    
def v_vertikal(v,theta):
    return(v*np.sin(theta))
```

```python
v[0] = v_vertikal(vy,sudut)
print(v)
ek[0] = energi_kinetik(v[0])
ep[0] = energi_potensial(y[0])
em[0] = energi_mekanik(ek[0],ep[0])
```

```python
for i in range(N):
    t[i+1] = dt*(i+1)
    v[i+1] = v[i] - dt*g
    y[i+1] = y[i] + dt*v[i+1]
    ek[i+1] = energi_kinetik(v[i+1])
    ep[i+1] = energi_potensial(y[i+1])
    em[i+1] = energi_mekanik(ek[i+1],ep[i+1])
```

```python
plt.plot(t,y)
plt.title('Posisi terhadap Waktu')
```

```python
plt.plot(t,v)
plt.title('Kecepatan terhadap waktu')
```

```python
plt.plot(t,ek)
plt.title('Energi Kinetik terhadap waktu')
```

```python
plt.plot(t,ep)
plt.title('Energi Potensial terhadap waktu')
```

```python
plt.plot(t,em)
plt.title('Energi Mekanik terhadap waktu')
```
