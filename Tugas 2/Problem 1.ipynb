```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
dt = 0.1
m = float(input('Massa = '))
g = 9.8
N = 100 #Partisi
v = np.zeros(N+1)
y = np.zeros(N+1)
t = np.zeros(N+1) 
ek = np.zeros(N+1)
ep = np.zeros(N+1)
em = np.zeros(N+1)
```

```python
v[0] = float(input('Kecepatan awal = '))
y[0] = float(input('Posisi awal = '))
t[0] = 0.0
```

```python
def energi_potensial(h):
    return (m*g*h)
def energi_kinetik(v):
    return (1/2*m*(v**2))
def energi_mekanik(kin,pot):
    return (kin + pot)    
```

```python
ek[0] = energi_kinetik(v[0])
ep[0] = energi_potensial(y[0])
em[0] = energi_mekanik(ek[0],ep[0])
```

```python
for i in range(N):
    t[i+1] = dt*(i+1)
    v[i+1] = v[i] + dt*g
    y[i+1] = y[i] - dt*v[i+1]
    ek[i+1] = energi_kinetik(v[i+1])
    ep[i+1] = energi_potensial(y[i+1])
    em[i+1] = energi_mekanik(ek[i+1],ep[i+1])
```

```python
plt.plot(t,v)
plt.title('Kecepatan terhadap Waktu')
```

```python
plt.plot(t,y)
plt.title('Posisi terhadap Waktu')
```

```python
plt.plot(t,ek)
plt.title('Energi Kinetik terhadap Waktu')
```

```python
plt.plot(t,ep)
plt.title('Energi Potensial terhadap Waktu')
```

```python
plt.plot(t,em)
plt.title('Energi Mekanik terhadap Waktu')
```
